## 1. Read a .csv file from a URL with PANDAS

```python
import pandas as pd 
```

```python
df_premier_21 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv')
# assigning a variable df_premier_21. 
```

```python
# showing first 10 rows dataframe 
df_premier_21.head(10)
```

```python
# rename column_name 
df_premier_21.rename(columns={'FTHG' : 'home_goal' , 'FTAG' : 'away_goal'}, inplace = True)
# syntax: table_name.rename(columns = {key : value, key : value, ...}, inplace = True)
# 'FTHG' : 'home_goal' is a key : value where current column name: FTHG is KEY and 
# renamed column name: home_goal is VALUE
# To update the column name we write inplace = True  where True is case sensitive 
```

```python
df_premier_21
```

```python
# Specify the folder path
folder_path = r'C:\Users\akrjha\Desktop\MC Scorecard\\'
# Save the DataFrame to a CSV file in the specified folder
df_premier_21.to_csv(folder_path + 'premier_21.csv', index=False)
```
